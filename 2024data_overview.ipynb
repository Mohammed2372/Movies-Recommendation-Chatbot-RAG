### check on string nulls

```python
sum = 0
for c in cols:
    null_count = df[c].isnull().sum()
    string_null_count = df[c].isin(['unknown', 'untitled', '', ' ']).sum()
    total_nulls = null_count + string_null_count
    print(f'columns {c} nulls: {total_nulls} (actual: {null_count}, string forms: {string_null_count})')
    sum += total_nulls
print(f'total nulls: {sum}')
```

too much nulls for the data

- drop useless columns (homepage, poster_path, backdrop, imdb_id)
- filter junk movies with vote_count less than 50
- handle missing dates (if year col is null, set it with 0)
- handle missing plot (if overview is null, replace it with empty string)

```python
## Check overview col

# Count overviews with less than 20 characters
short_overview = df['overview'].str.len() < 10
short_count = short_overview.sum()
print(f'Total overviews with less than 20 chars: {short_count}')
print(f'Percentage: {(short_count / len(df)) * 100:.2f}%')

# Print the problematic overviews
print('\nProblematic overviews (less than 20 chars):')
print('='*10)
for idx, (i, row) in enumerate(df[short_overview].iterrows(), 1):
    overview = row['overview'] if pd.notna(row['overview']) else '[NULL]'
    title = row['title']
    print(f"{idx}. Title: {title}")
    print(f"   Overview: '{overview}' (length: {len(str(overview))})")
    print()
```

there are some wrong titles with strange overviews that need to cleaned
