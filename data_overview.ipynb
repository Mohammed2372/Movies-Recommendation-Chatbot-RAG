## status col


```python
df.status.unique()
```

```python
for i in df.status.unique():
    print(f'status {i} count is: {(df.status == i).sum()}')
```

```python
status_counts = df['status'].value_counts()
fig, ax1 = plt.subplots(1, figsize=(14, 5))

# Bar chart
ax1.bar(status_counts.index, status_counts.values, color=['#e74c3c', '#2ecc71', '#f39c12'])
ax1.set_ylabel('Number of Movies')
ax1.set_xlabel('Status')
ax1.set_title('Movie Count by Status')
ax1.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## runtime col


```python
df.runtime.unique()
```

```python
runtime_gte_60m = (df['runtime'] >= 60) & (df['runtime'] < 90)
runtime_gte_90m = (df['runtime'] >= 90) & (df['runtime'] < 120)
runtime_gte_120m = (df['runtime'] >= 120) & (df['runtime'] < 150)
runtime_gte_150m = df['runtime'] >= 150
```

```python
print(f'Movies runtime >= 1 hour and < 1.5 hours: {runtime_gte_60m.sum() / df.runtime.count() * 100:.2f}%')
print(f'Movies runtime >= 1.5 hours and < 2 hours: {runtime_gte_90m.sum() / df.runtime.count() * 100:.2f}%')
print(f'Movies runtime >= 2 hours and < 2.5 hours: {runtime_gte_120m.sum() / df.runtime.count() * 100:.2f}%')
print(f'Movies runtime >= 2.5 hours: {runtime_gte_150m.sum() / df.runtime.count() * 100:.2f}%')
```

```python
runtime_categories = ['60-90m', '90-120m', '120-150m', '150m+']
runtime_counts = [runtime_gte_60m.sum(), runtime_gte_90m.sum(), runtime_gte_120m.sum(), runtime_gte_150m.sum()]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Pie chart
ax1.pie(runtime_counts, labels=runtime_categories, autopct='%1.1f%%', startangle=90)
ax1.set_title('Distribution of Movie Runtimes')

# Bar chart
ax2.bar(runtime_categories, runtime_counts, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
ax2.set_ylabel('Number of Movies')
ax2.set_xlabel('Runtime Category')
ax2.set_title('Movie Count by Runtime Category')
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

## vote_average col


```python
df['vote_average'].unique()
```

```python
vote_ste_5 = df['vote_average'] < 5
vote_gte_5_sm_7 = (df['vote_average'] >= 5) & (df['vote_average'] < 7)
vote_gte_7 = (df['vote_average'] >= 7)
```

```python
print(f'rate average less than 5 stars: {vote_ste_5.sum() / df.vote_average.count() * 100:.2f}%')
print(f'rate average greater than 5 stars: {vote_gte_5_sm_7.sum() / df.vote_average.count() * 100:.2f}%')
print(f'rate average greater than 8 stars: {vote_gte_7.sum() / df.vote_average.count() * 100:.2f}%')
```

```python
vote_categories = ['< 5 stars', '5-7 stars', '>= 7 stars']
vote_counts = [vote_ste_5.sum(), vote_gte_5_sm_7.sum(), vote_gte_7.sum()]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Pie chart
ax1.pie(vote_counts, labels=vote_categories, autopct='%1.1f%%', startangle=90)
ax1.set_title('Distribution of Movie Ratings')

# Bar chart
ax2.bar(vote_categories, vote_counts, color=['#ff6b6b', '#ffd93d', '#6bcf7f'])
ax2.set_ylabel('Number of Movies')
ax2.set_xlabel('Rating Category')
ax2.set_title('Movie Count by Rating Category')
ax2.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
